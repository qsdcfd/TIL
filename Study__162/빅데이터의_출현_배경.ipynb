### 빅데이터의 출현 배경

- 기업의 고객 데이터/트래킹 수집행위 증가

   - 기업들의 온라인/오프라인 수집 정보, 소비자 형태에 대한 정보 수집 가능
   
   - 고객관련 정보 수집의 증가로 더 많은 데이터 스트리지와 정교한 분석 능력 필요

<br>

- 멀티미디어 콘텐츠의 콘텐츠 사용에 관한 정보 증가

   - CT스캔, CC카메라등 다양한 부분에서 대용량 멀티미디어 콘텐츠 생산 증가
   
   - 고화질 동영상은 이미 인터넷 전체 트래픽의 50%이상을 차지하고 있음
   
   - 오리지널 콘텐츠뿐만 아니라 콘텐츠 소비에 관한 정보도 대량 생산(사용자 정보, 선호등)
   
<br>

- SNS의 급격한 확산과 비정형 데이터의 폭증

   - 트위터, 페이스북 등 sns의 급격한 확산과 더불어 텍스트 등 비정형 데이터의 폭증이라고 할 수 있습니다. 
   
   - SNS에서 유통되즌 정보는 비정형 데이터로서 이의 처리를 위해서 추가적인 데이터 처리가 필요하게 되어 데이터의 복잡성이 증가했습니다
 
<br>


- M2M및 IoT확산에 따른 센서 저변 확대

   - M2M(머신투머신) 및 IoT(인터넷 of Things) 등의 통신 기술의 발전에 따라 사물간 통신 네트워크에서 발생하는 데이터가 급격히 증가하게 되었습니다.

   - M2M, IoT 등이 활성화 되면서 데이터를 사용자가 생성하지 않고, 인프라 자체가 다량의 데이터를 생성하게 되었습니다.
   
   - 위와 같은 대용량의 데이터를 빅데이터라고 부릅니다.
   
   
<br>

### 각 업계별 출현 배경

**빅데이터의 현상은 없었던 것이 새로 등장한 것이 아닌 기존의 데이터 ,처리방식, 다루는 사람과 조직 차원에서 일어나는 변화**

- 산업계

   - 고객 데이터 축적 양이 증가
   
<br>

- 학계

   - 거대 데이터를 확용한 연구 확산
   
<br>

- 기술업계

   - 디지털화, 저장기술, 인터넷보급, 모바일 혁명, 클라우드컴퓨팅 관련 기술의 발전
   
   
<br>

### 빅데이터 출현으로인한 이슈

- 빅데이터 가치에 대한 실증적 접근

   - 스마트 시대에선 소셜, 사물, 라이프로그 데이터 등이 결합되어 빅데이터의 영향력 증가
   
   - 실시간의 연결과 소통의 스마트혁명으로 데이터가 폭증하게 되어서 기존의 데이터 저장 관리 분석 기법은 한계와 도전에 직면 
   
   - 빅데이터로 환경 변화는 정보 사회의 패러다임을 견인할 정도의 큰 힘을 발휘하기에 빅데이터 활용의 수준이 경쟁력과 성패를 좌우

- 프라이버시 침해 급증

- 인력수급의 불균형 심화

- 빅데이터의 실시간성과 빅데이터 처리과정의 업그레이드

### 빅데이터의 기여

- 기술 경쟁력

빅데이터 관련 기술연구개발 지원과 데이터 과학자등ㄹ 전문 인력 양성을 통한 기술 경쟁력 확보해야한다.

- 생산성 향상

민간 공공에서 측정되는 빅데이터의 활굥으로 생산성을 향상시키고 비용을 대폭 절감시킵니다.

- 정부 혁신

빅데이터 활용으로 사회현안 및 국민의 니즈파악으로 미래 전력을 수렵하고 공공서비스 제공합니다.
