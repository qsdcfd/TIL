# 이미지로 연습하는 numpy

`PIL` 모듈은 이미지를 제어할 수 있는 기능을 제공합니다. `open` 함수로 저장된 이미지 데이터를 읽어 올 수 있습니다. img에는 `PngImageFile` 객체가 저장됩니다. 

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from PIL import Image
img = Image.open("/content/drive/MyDrive/Colab Notebooks/2022-04-18/3.numpy/image/0.png")
img = img.convert('L')
img
```

- L : 흑백
- RGB : Red, Green, Blue 삼원색을 이용한 색 표현

`transpose` 메서드로 `flip`할 수있음

```python
img.transpose(method=1)
```

`PngImageFile` 객체를 `ndarray`로 변환

```python
import numpy as np

arr = np.array(img)
print(type(arr))
print(arr.shape)
```

```python
arr
```

ndarray는 matplot 라이브러리로 시각화할 수 있습니다. 

```python
from matplotlib import pyplot as plt
plt.imshow(arr, cmap='gray')
```

Q. 이미지 Crop하기 (200, 200) 크기로 영상 자르기

```python
arr[   :200 ,  :200  ]
```

```python
plt.imshow(arr[  100 :300 ,  100:300  ], cmap='gray')
```

200보다 큰 값을 255로 변환하라. 

```python
t = arr.copy()    # 원본이 훼손되지 않게 복사본 생성
t[t>200] = 255
plt.imshow(t, cmap='gray')
```

동일한 영상을 가로 방향으로 이어 붙여 출력하라.

```python
temp = np.hstack(  (arr, arr) )
plt.imshow(temp, cmap='gray')
```

영상 회전 (90/180/270/360)  
`np.rot90`

```python
plt.imshow(np.rot90(arr), cmap='gray')
```

영상의 반전

```python
plt.imshow(255-arr, cmap='gray')
```
