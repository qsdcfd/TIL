### 바이오헬스 산업의 국내외 상업화 사례

#### 국내 상업화 사례

**세포치료제분야**

1.안트로젠의 크론병 누공 치료제 큐피스템

크론병이란 입에서 항문까지 소화관 전체에 걸쳐 발생하는 만성 염증성 장질환입니다. 면역 불균형과 직접적으로 연관하여 염증이 유발되는 것으로 알려져 있습니다.

큐피스템은 자가 지방유래 중간엽줄기세포로 만들어진 줄기세포치료제인데요. 
2012년에 국내 식약처 허가를 받았고, 2014년부터 발매하기 시작했습니다.

큐피스템 투여가 이루어진 부위는 면역체계를 리밸런싱하기 때문에, 기존의 대증적 치료와는 다른 근본적인 치료가 가능합니다.

2.파미셀의 급성심근경색증 줄기세포치료제 에이엠아이입니다. 

에이엠아이는 자가골수유래 중간엽줄기세포를 이용하여 제조한 치료제인데요.

흉통 발현 후 72시간 이내에 관상동맥성형술을 시행하여 재관류된 급성심근경색 환자에서의 좌심실구혈률을 개선하는 효능이 있습니다.

3.테고 사이언스의 배양피부 칼로덤

테고 사이언스의 배양피부 칼로덤은 우리나라 최초의 동종유래, 즉 타인의 세포를 원료로 사용한 배양피부입니다.

깊은 2도 화상과 당뇨성 족부궤양에 적용하면 상처를 빨리 아물게 하여 치료기간을 단축시키고 흉터생성을 최소화하는 기능을 가진 치료제입니다.

<br>

**바이오 시밀러**

바이오 시밀러는 동등생물의약품이라는 용어로 특허가 만료된 생물의약품에 대한 복제약을 말합니다.
바이오 시밀러 가운데 셀트리온의 세 가지 치료제에 대해 알아보도록 하겠습니다.

1.혈액암 치료용 항체인 바이오시밀러 트룩시마입니다. 

바비오시밀러 트룩시마는 2016년 11월 국내에서 판매허가를 받았는데요.
세포의 과도한 성장과 과도한 확산을 억제하는 효능이 있어서 비호지킨 림프종, 만성림프성백혈병, 류마티스 관절염, 다발성 혈관염이 있는 육아종증 및 현미경적 다발혈관염 환자의 치료에 사용됩니다.

2.자가면역치료제, 램시마입니다.

램시마는 세계 최초의 단일클론 항체 바이오시밀러 의약품인데요, 크론병(CD), 궤양성대장염(UC), 류마티스관절염(RA), 강직성척추염 (AS), 건선성관절염(PsA) 및 판상건선(PsO)을 비롯한 다수의 중증 자가 면역 질환 환자들의 치료에 사용되고 있습니다.

3.유방암 치료제 허쥬마입니다. 

허쥬마는 2017년도 하반기부터 국내 발매예정인 바이오시밀러 의약품으로 전이성 유방암과 조기 유방암, 전이성 위암 치료에 사용될 예정이라고 합니다.

삼성 바이오에피스의 자가면역질환 치료제 브렌시스는 류마티스 관절염, 건선성 관절염, 축성 척추관절염(강직성 척추염, 방사선상으로 확인되지 않는 축성 척추관절염), 건선 치료제 등으로 쓰입니다.

#### 국외

1.투애니쓰리 앤드 미의 블룸증후군테스트입니다.

이 블룸증후군 부모로부터 자녀에게 대물림될 가능성이 있는지를 검사하기 위한 목적으로 개발되었습니다.

블룸증후군은 희귀유전 질환으로 단식, 광민감도 증가, 얼굴 모세혈관학장증, 면역 결핍에 의한 감염 감수성 증가, 암에 대한 감수성이 크게 증가하는 특징이 나타나는 병입니다.

미국에서는 일반인들이 가정에서도 투애니쓰리 앤드 미의 테스트를 통해서 손쉽게 스스로의 블룸증후군 열성 유전 형질의 보유 여부를 검사 할 수 있다고 합니다.

2.아아헬쓰의 얼라인 스마트폰 글루코미터입니다.

얼라인 스마트폰 글루코미터는 당뇨환자들이 스마트폰과 헤드폰 잭으로 손수 비게 혈당을 측정할 수 있도록 하는 것으로, 아이폰의 ‘헤드폰 잭’에 연결해 사용할 수 있습니다.

기존의 혈당 측정기를 사용하는 방법과 유사합니다. 측정 결과는 앱을 통해 보여지고, 일정 기간 동안의 측정치와 추이가 확인 가능합니다.

또한, 측정 결과는 앱에 저장되어 일일이 혈당치를 기록하는 불편이 없다는 장점을 지닙니다.

3.센시미드의 트리거피쉬입니다.

트리거피쉬는 녹내장 환자의 안압을 24시간 측정하는 콘택트렌즈인데요, 렌즈 내부에 장착된 센서와 안테나를 통해 안압을 측정하고 측정한 데이터를 블루투스를 통해 의사의 컴퓨터로 전달되는 시스템입니다.

진단 방식은 눈 주위에 놓인 센시미드 트리거피쉬 안테나가 콘택트렌즈에서 정보를 무선으로 수신하고 수신한 데이터는 얇은 플렉시블 케이블을 통해 안테나에서 휴대용 레코더로 전송됩니다.

환자가 착용한 휴대용 레코더는 모니터링 세션 중에 획득한 데이터를 저장하는데 기록 기간이 끝나면 데이터는 블루투스를 통해 레코더에서 의사의 컴퓨터에 설치된 소프트웨어로 전송됩니다.

4.칼디오베스큘러(Cardiovascular)의 어움(AUM)입니다. 

어움(AUM)은 심장박동 소리의 변화를 통해 동맥질환의 징후를 진단하는데 관상 동맥이 폐색될 경우 발생하는 독특한 소리를 판별하여 관련 질환을 사전에 진단 및 예방합니다.

어움(AUM)만으로는 정확히 어떤 혈관이 폐색되었는지는 알 수 없으나, 낮은 가격으로 정밀검사 유무를 결정 가능하다는 장점이 있다고 합니다.
