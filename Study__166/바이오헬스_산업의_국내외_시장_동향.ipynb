
#### Intro

4차 산업혁명은 AI나 IoT, 빅데이터 등이 핵심기술이 되는 사회인데요, 바이오헬스산업도 4차 산업혁명의 핵심 분야로 떠오르면서 전 세계적으로 관심이 집중되고 있습니다.

또한, 경제협력개발기구인 OECD에 따르면 2030년에는 바이오산업이 경제 성장을 이끄는 핵심 동력으로 자리 잡는 ＇바이오경제 시대＇가 될 것이라고 말하고 있습니다.

<br>

#### 국내

한국의 바이오헬스산업은 크게, 생명공학 기술 기반 바이오산업, 제약산업, 의료기기산업 등으로 크게 분류할 수 있습니다.

*바이오산업

바이오헬스산업은 약 30조 원으로 한국 GDP의 약 1.9%를 차지하는 것으로 나타났습니다.

바이오산업에는 바이오 의약산업, 바이오 화학산업, 바이오 식품산업, 바이오 환경산업, 바이오 전자산업, 바이오 공정 및 기기산업, 바이오 에너지 및 지원산업, 바이오 검정, 정보서비스 및 연구개발산업 등으로 나누어 볼 수 있습니다.

그 중 가장 큰 비중을 차지하고 있는 분야는 바이오의약산업과 바이오식품산업 으로 전체 바이오산업의 약 70%의 비중을 차지하고 있는 것으로 나타나고 있습니다.

이를 위해서는 국내 바이오헬스산업의 전체 시장규모 확대를 위한 방안에 관심을 가져야 할 시점이 왔다고 볼수 있습니다.

<br>

*제약산업


국내 바이오의약품 시장 규모를 살펴보면, 2014년에 약 1조 9천 8백억원으로 전체 의약품 시장(약 19조원)의 약 10.25%로 낮은 비중입니다.

그러나 성장률은 매우 높게 나타나고 있는 것을 알 수 있었습니다. 바이오 의약품 산업의 예로는 단백질 의약품, 항체 의약품, 백신, 유전자 치료제 등이 있습니다

<br>

*의료기기 산업

국내 의료기기 산업 시장의 현황을 살펴보면, 수출은 25.8억 달러로 9.3%, 수입은 29.3억 달러로 7.3%가 증가한 것으로 보이고 있습니다.
2014년 의료기기 시장규모는 4조 9천7백5십억 원으로 전년도 대비 7.4% 증가한 것으로 나타나고 있습니다.

이렇게 의료기기 시장의 규모가 점차 증가하는 이유는 의료 패러다임이 치료에서 예방·진단 중심으로 변화하면서, 건강하고 행복한 삶에 대한 요구로 의료기기 수요가 증대함에 따라 국내 의료기기 생산은 꾸준한 성장세를 유지하고 있는 것으로 볼 수 있습니다.

<br>

국내 줄기세포 및 재생의료 분야의 시장 규모에 대해 알아보겠습니다.

전 세계에서 우리나라 줄기세포 시장은 약 2%대의 점유율을 차지하고 있는데요. 국내 줄기세포 시장은 2005년에서 2015년까지 25%의 성장률을 보이고 있습니다.

또한 한국에서는 총 64건의 줄기세포치료제 임상시험이 이루어졌고, 아시아권에서는 1위로 전 세계적으로는 미국에 이어 2위로 나타나고 있습니다.
한국이 허가한 네개 제품의 줄기세포치료제 중 FCB파미셀의 ‘하티셀그램-AMI’는 세계 최초로 줄기세포치료제로 품목 허가를 받았습니다.

<br>

#### 국외

*의료 기기 산업

2014년 국외 의료 기기 시장의 규모는 북미와 남미 지역이 1546억 달러로 세계시장의 약 45.4%로 가장 큰 규모를 보였는데요.

그 뒤를 이어서 서유럽 990억 달러로 전체 25.9%, 아시아태평양 723억 달러로 21.2%, 중부 및 동유럽 171억 달러 5%, 중동/아프리카 83억 달러로 2.4% 순으로 나타났습니다.

<br>

2019년 시장 성장의 주요 요인은 고령화 추세, 건강에 대한 관심 고조 및 웰빙에 대한 사회적 분위기 확산, 주요 국가들의 보건의료 정책, 브릭스 등의 경제 성장으로 인한 의료 서비스 수요 증가 등이 있습니다.

지역별로는 아시아태평양 지역이 2019년에 천백십육억 달러로 연평균 9.1% 성장할 것으로 전망되구요.
중동/아프리카 지역의 시장 규모는 작으나 성장률은 10.3%로 높을 것으로 전망되고 있습니다.
또한 북미와 남미 시장은 2019년 2천백삼십구억 달러로 45.7%를 점유하고 있고, 서유럽 시장은 천오십육달러로 22.6%, 중앙 및 동유럽은 232억 달러로 5.0% 비중이 될 것으로 추정하였습니다.

<br>

*줄기세포 시장

줄기세포 시장은 2013년에 400억 달러 규모를 이루었으나, 향후 지속적으로 성장해서
2018년까지 1,177억 달러 규모로 성장할 것이라 전망되고 있습니다.

또한 줄기세포치료제 및 치료기술 관련 임상시험의 증가와 줄기세포은행 서비스 산업활성화 추세인데요
이에 따라서 글로벌 줄기세포 시장은 연평균 성장률 2013년부터 2018년까지 약 24%로 확대될 것으로 전망되고 있습니다.

전 세계 지역별 줄기세포 시장규모를 살펴보겠습니다.

북아메리카 지역이 54% 약 216억 달러로 가장 큰 비중을 차지하고 있는 것을 알 수 있었습니다.
그 뒤로 유럽 24%로 96억 달러, 아시아-태평양 14%로 56억 달러 순으로 시장규모가 형성되어 있습니다.
