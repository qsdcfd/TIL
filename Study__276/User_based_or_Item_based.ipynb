## Collaborative Filtering

### Neighborhood-based Collaborative Filtering

- Memory-based CF라고 불린다.

- User-Item간의 평점등 주어진 데이터로 새로운 아이템을 예측한다.

- 특징

   - 구현이 간단하다

   - Model-based CF에 비해 계산량이 적다

   - 새로운 user,item이추가되더라도 비교적 안정적이다

   - 새로운 content(user or item) 추천

   - NN개념과 유사하다

### User-based Collaborative Filtering

- 두 사용자가 얼마나 유사한 항목을 좋아했는지 바탕으로 추천한다.

   - 취향이 비슷한 사용자끼리의 데이터를 바탕으로 추천한다.

- 플랫폼에서 유저가 많은 부분을 차지하는 시스템에 적용하기에 적절한 알고리즘이다.

- SNS의 친구 추천 등에서 활용이된다

- Regression vs Classification

   - 평점 등 점수 분포가 continuous일 때와 discrete일 때 나눠서 적용

   - 평점등 점수를 정규화하여 추천 알고리즘에 적용

      - Mean-Centering, Z-score normalization

<br>

### Item-based Collaborative Filtering

- 아이템과 아이템 사이의 유사도 계산

- 과거 아이템 선호도 데이터를 기반으로 선호 연관성 높은 다른 아이템을 추천한다.
   - 컨텐츠기반추천시스템과 차이점은 여러 사용자의 과거 선호도 데이터로 연관성 높은 아이템을 찾아서 해당 아이템을 사용자에게 추천

- 아마존, 넷플릭스 등에서 많이 활용된다.

### User-based vs Item-based Collaborative Filtering

- 정확도

   - User수 < item수: User-based
   - User수 > item수: item-based

<br>

- 모델 Robustness

   - 유저와 아이템이 얼마나 자주 그리고 많이 변하는지에 따라서 다르다

   - 아이템 수가 크게 변하지 않는다면, item-based다

<br>

- 설명력

   - Item-based는 비슷한 item과 가중치로 함께 설명이 가능하다

   - User-based는 특정 user와 비슷한 user로 분류된 user의 실제취향은 알기 어렵다

<br>

- 새로운 추천 가능

   - Item-based는 과거 아이템에 의존하기 때문에 새로운 item을 추천하기는 어렵다

   - User-based는 여러 유저의 데이터를 보기 때문에 더 새로운 추천을 할 수 있다.

### User-based & Item-based Collaborative Filtering

- Cold-start문제: 딥러닝 기반 추천 알고리즘

   - 충분한 데이터가 없다면 좋은 추천을 할 수 없다
   - 유저에 대한 아무런 기록이 없다면, 새로운 아이템에 대한 정보가 없다면, 추천이 불가능하다

<br>

- 계산량

   -  데이터가 많아질수록 유사도 계산이 많아진다.
   - 하지만 데이터가 많아야 추천 품질이 좋아진다.

<br>

- Long-Tail Economy:모델 기반 협업 필터링

   - 대부분의 사용자가 관심 갖는 소수 아이템으로 쏠림 현상이다.

   - 관심이 상대적으로 부족한 아이템은 추천되지 못하는 현상
