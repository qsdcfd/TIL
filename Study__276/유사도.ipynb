### 자카드 유사도 - Jaccard Similarity

- 집합의 개념을 이용한 유사도 계산으로 집합 A와 B사이의 유사도는 얼마나 많은 아이템 혹은 element가 겹치는지 판단

- 교집합/ 전체집합의 비율로 유사도 계산

- 공통 아이템이 없으면0 , 모두 겹치면 1

- 주로 여러 단어로 구성된 등 문서 혹은 문장이 유사한지 판단할 때 사용

### 피어슨 유사도- Pearson Similarity

- Vector X,Y 사이의 상관계수를 계싼(=피어슨 상관계수)

- 각 Vector의 표본평균으로 각 vector를 정규화하고 코사인 유사도를 구한다

- 1은 양의 상관관계 , -1은 음의 상관관계, 0은 상관관계없음
