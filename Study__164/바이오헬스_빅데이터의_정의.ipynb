### 바이오헬스

*정의

바이오헬스란 생명시스템의 다양한 생체상태에 대한 생체정보라든가 뇌과학, 그리고 의료기기, 또 ICT 융합을 기반으로 동작원리를 이해하고

이해한 동작원리를 제어하여 생체상태를 식별하거나 선제적으로 예방하거나, 맞춤 관리 및 질병 진단을 하거나, 또 치료에 활용하는 차세대 스마트 의료기술

<br>

*산업

*산업 분류

- 맞춤 의료

   - 맞춤의료는 질병을 예방하고 진단하고, 치료하는것과 관련된 의사결정을 내려주는 도구입니다.
   
   - 개인들의 개별적인 유전적 프로필을 사용해서 새로 구상하는 의학적 실천을 말합니다.

   - 핵심 기술: 유전체 연구
   
   - 핵심 산업 모델: 생체 분석 서비스, 바이오칩, 맞춤형 표적 치료제, 맞춤 의료 서비스

<br>

- 재생 의료

   - 재생으로는 손상된 세포와 조직 그리고 장기를 대체하거나 재생시켜서 정상기능을 복원시키는 의학 분야를 말합니다.
   
   - 회복이 불가능했던 과거에는 조직이나 장기들의 회복 매커니즘을 활성화 시키거나 교체를시켜서 손상된 부위를 재생하는것을 의미하는데요
   
   - 자가치유가 불가는한 장기를 실험실에서 배양하고 안전하게 이식하려는 시도들도 포함됩니다.
   
   - 핵심 기술: 줄기세포 연구
   
   - 핵심 산업 모델: 세포 치료, 조직 재생, 바이오 장기
   
<br>   

- 바이오 의약
 
   - 바이오 의학은 사람이나 다른생물체에서 유례된 물질을 원료나 재료로 이용해서 생명공학 기술을 이용해 만드는 기술을 총칭하는것입니다.
   
   - 핵심 기술: 유전자 제조합기술
   
   - 핵심 산업 모델: 항체치료제, 바이오 시밀러, 비정형 데이터
   
<br>

### 바이오헬스 빅데이터

*개념

![](https://images.velog.io/images/qsdcfd/post/0dbc4c20-b755-4a1a-a171-f2291f93f65e/image.png)

빅데이터 기반 창출의 방법을 활용하여 방대한 바이오 정보를 효율성 있게 생물 및 의학적인 의미를 밝혀서 새로운 지식 및 서비스 개발

<br>

*등장 배경

- 2,000년에 완성된 인간 지놈프로젝트 이래 현재까지 축적된 엄청난 유전체 정보가 축적되고 생성되면서 의미 있는 정보분석을 도출할 필요성이 증대


- 아울러 의료서비스는 확대되고 있고 이에 바이오 산업육성을 통해 관련분야의 발전을 모색하기 시작

- 맞춤형 의료서비스에 대한 관심도 증가

- 진료기록과 의료 영상뿐만 아니라 유전자 통계와 전염병 현황 등을 포함한 방대한 보건 의료 정보가 늘어나고 있고, 그 중에서 유의미한 정보를 얻어내는 연구의 활성화가 요구되었기 때문입니다
