### 빅데이터의 활용 사례

**의료 건강 분야**

- 아비바 생명 고객 맞춤형 보험 상품"RateMyDrive"

   - 운전자의 운전 패턴에 기반한 모델 상품
   
   - 주로 운전하는 시간과 지역을 감안하여 보험료 산정
  
   - 가입자들에게 호평을 받음
   
<br>

- 건강보험심사평가원 "실시간 국민관심 질병예측 서비스"

   - 대국민 서비스: GIS기반의 대국민 건강지도 서비스 제공하여 치료에서 예방 중심, 고령화 사회에 따른 의료비 절감 및 국민 건강 증진
   
   - 기관 연계 서비스: 질병관리본부 등 대외 기관에서 국민관심질병 예측 지수등 제공
   
   - 언론/ 방송 연계 서비스: 언론매체를 통해 지속적 질병예방 활동 지원

<br>

**경제-경영분야** 

- 아마존닷컴 추천 상품

   - 모든 고객들의 구매 내역을 데이터베이스에 기록하여 분석한 후 소비자의 소비 취향과 관심사를 파악하고 있습니다.
   
   - 위의 방식으로 모은 빅데이터를 활용하여 고객별 추천상품 표시
   
<br>

- 플럭 터빈 제조 설비업체 "베스타스" 

   - 날씨, 조수간만의 차, 위성 이미지등 페타바이트 규모의 정형 데이터와 비정형 데이터를 수집
   
   - 실시간으로 변하는 기후 요소 데이터를 수집하고 분석한 후 최적의 풍력 발전소 부지 선정
   
   - 풍력 터빈이 얼마나 풍력 에너지를 생산할지 전망
   
   - 기존에 공급한 풍력 터빈의 유지 보수 일정을 위한 최적의 시간 계산
   
<br>

**농업분야**

- 후지쯔의 농업용 빅데이터 분석 솔루션

   - 기후와 토양 환경들을 자동으로 체크하는 IoT활용
   
   - 최적의 파종, 필요한 농약의 양, 수학 시점등을 분석
   
   - 농지작업 실적과 작물 이미지등의 데이터를 분석하여 수확량 증가와 품질을 향상시키는 클라우드 기반의 빅데이터 분석 솔루션
   
   - 농업 종사자들의 합리적인 작업 기능을 가능하게 함
   
<br>

**치안분야**

- 캔프스탯

   - 뉴욕시의 과거지향적인 범죄분석 서비스
   
   - 모든 경찰서로 한 주에 벌어진 범죄정보를 제공하여 범죄 분포상태를 알 수 있고 이를 지도로 시각화하여 범죄 대응전략 수집하는 시스템
   
 <br>
 
 **환경분야**
 
 - 영국 환경청 실시간 홍수 경고지도 서비스
 
   - 지역별 홍수 발생 가능성을 예측하여 실시간으로 지도에 표시하는 실시간 홍수 경고지도 서비스 제공
   
   - 홍수 발생 가능성에 따라 심각한 홍수 정보, 홍수 경보, 홍수 경계, 홍수 영향에서 벗어나 총 4단계로 구분
   
   - 위험지역을 도출 뒤 결과를 웹으로 통해 배포
   
   - 홍수 피해 위험 예상 지역 도출 및 홍수 대비 방안 수립 가능
   
   
<br>

**문화분야**

- 영화 추천서비스 왓챠

   - 한국의 벤처 회사에서 개발한 영화 추천 서비스
   
   - 가입 후 30개의 영화 별점 평가를 매기며 사용자의 취향을 분석하고 비슷한 취향 추천
