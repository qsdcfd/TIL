## Item Profile

1. 모든 아이템에 대해서 profiling

   - Feature extraction

2. Item Profiling = Vector Representation

   - One-hot vector
   - Pandas dataframe-dummy vector
   - etc..

3. Sample features

   - 영화,작가, 제목, 배우 등
   - 글: 중요한 단어, TF-IDF

<br>

### Item Representation

- Item을 set of features로 표현한다.

- Item의 여러 특성을 attributes, features라고 한다

- Item Profile는 분석된 여러 특성들을 나타내어 정령화한다.

   - 구성된 목록등을 바탕으로 가중치 부여

- Item 설명은 text features로, text preprocessing등을 활용해여 features를 얻는다.

   - 설명등 문서가 포함하는 키워드 활용
   - TF-IDF등으로 가중치 부여

- Item-Item간의 유사도는 Cosine Similarity활용

<br>

**활용**

- Features Extraction

   - 유저의 관심사를 나타내는 많은 정보를 얻는다면 추천 결과가 만족스럽다
   - Feature Weighting 또는 Feature Selection을 통해서 discriminative한 features추출


- 영화

   - Meta information:영화 장르, 배우, 감독, 간단한 설명, 예고편

   - 영화 리뷰에 대한 키워드 분석, 영화 설명 키워드 분석

- 음악 

   - Meta Information:음악 장르, 음악 가사, 음원 데이터
   - 음악 리뷰, 함께 소개된 음악 종류, 제목 등

- 뉴스, 웹페이지 등

   - 기사 원문, 헤드라인, 댓글 등 텍스트 정보




## User Profile

- 데이터로부터 유저 성향 파악

   - User ratings, implicit feedback

- 사용자가 가지는 아이템들의 특성 또는 조사된 사용자 특성으로 구성

- 특성 가중치는 사용자가 가지고 있는 아이템 특성 가중치의 평균 값을 주로 활용한다.

- 사용자의 기호를 따라 파악하거나, 설문조사 혹은 자발적인 키워드 입력등으로 얻은 프로필

   - Explicit and Implicit feedback to get user profile

- Like/DistLike, Ratings from 0 to 5, 직접 작성한 유저의 주관적인 생각을 담은 댓글

- 검색 로그, 아이템 선택 후 구매 혹은 선택, 특정 아이템을 위해 비슷한 아이템을 tag한 이력

- User Profile학습은 분류 문제를 푸는 것과 같다.

   - 주어진 user profile정보를 바탕으로 새로운 아이템 Inew을 유저가 평가한 metrice 중 하나 선택

   - 아이템에 대한 유저의 기존 평가 = {r1,r2,..rs}
