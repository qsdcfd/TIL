### 모수의 통계량

**모수(parameter)**

모집단의 특성을 나타내는 값으로 알 수 없는 상수인 모수의 값을 추정하기 위해서 모집단으로부터 크기 n인 표본 X1,X2,...Xn을 추출해야한다.

<br>

**통계량(statistic)**

표본들만의 함수로 표현하는 것으로 특정 모수를 추정하기 위해서 표본을 이용하여 만든 함수입니다.

![](https://images.velog.io/images/qsdcfd/post/5632547d-56cc-41fe-ba13-d355329283c8/image.png)

<br>

**표본분포(sampling distribution)**

통계량이 갖는 분포

<br>

### 중심극한 정리


![](https://images.velog.io/images/qsdcfd/post/5f08eaea-aa3e-4d4f-b845-9761bb0b6ce6/image.png)


표본의 크기가 클수록 X바의 분포는 더 정규 분포 가까이에 수렴하기에 표본의 크기가 25보다 클 경우 중심극한정리를 이용할 수 있다

```
x=floor(runif(2500,0,10)); x  
    # 0부터 9까지의 정수를 균일분포로 2,500개 생성
```

![](https://images.velog.io/images/qsdcfd/post/58ce5616-8f3c-4667-8a02-986db55250dc/image.png)

```
hist(x)

```

![](https://images.velog.io/images/qsdcfd/post/cb967405-b71f-4da4-b308-643c0ebc2ac6/image.png)

```
mean(x)  # E(X) = 4.5와 유사함

[1] 4.5588
```

```
sd(x)  # s.d.(X)= 2.87와 유사함
[1] 2.877126
```

```
y = array (x, c(500,5)) ;y  
# 2,500개의 정수를 크기가 5인 벡터 500개를 생성
```

```
xbar = apply(y,1,mean); xbar
    # 크기 5인 표본에서 표본평균을 구하여 500개의 표본평균 생성 
```

```
hist(xbar)
```
![](https://images.velog.io/images/qsdcfd/post/91975c25-ac62-40fc-a69b-cdbcbe818c97/image.png)


```
 mean(xbar)  # 표본평균의 기댓값으로 E(X) = 4.5와 유사함
[1] 4.516
```

```
sd(xbar)  # 표본평균의 표준편차로서 s.d.(X) /√(5 )= 1.28와 유사함
[1] 1.283401
```
