## 스택, 큐


### 스택(LIFO)

- Last In first Out

- push():요소 추가

- pop(): 가장 최근 삽입 요소 제거



- 콜스택: 컴퓨터 프로그램의 서브 루틴 정보 저장

- 스택 오버플로우: 꽉 찬 스택에 요소 넣기

- 스택 언더플로우: 없는 스택에 요소 빼기


*우리의 생각(좌) . 연결리스트(우)



#### 연결 리스트를 이용하여 스택 ADT구현

*리스트를  담을 None클래스



*Stack 클래스



- 초기화 함수 __init__()

   - 노드 값: item
   
   - 다음 노드 포인터: next

   - pop(): last포인터를 한 칸 앞으로 지정하여 마지막 아이템 꺼내는 추가된 값을 가리킨다.

### 큐(FIFO)

- First in First Out

- 줄을 서는 것과 비슷합니다.

- 활용도가 낮다(스택에보다)

- 데크, 우선순위 큐, 너비 우선 탐색나 캐시에서 유용하게 쓰입니다.

- 리스크가 큐의 모든 연산을 지원하지만, 여러 분야에서 O(1)이 가능한 데크가 좋긴합니다.
