**변수(Variables)**

원하는 데이터를 넣을 수 있는 공간을 말합니다.

메모리 할당이 가능합니다.

변수의 이름을 마음대로 지정이 가능하지만, 최대한 간결하게 의미를 갖는 것을 설정하는 것이 좋다.

x,y는 메모리에 할당된 변수의 이름

= : 할당자이고 해당 변수에 특정 값을 넣어주는 역할입니다.

12.2,14,100: 값이고 해당 변수에 우리가 넣은 값입니다.

```
x = 12.2# 변수 x에 12.2할당
print(x)# 12.2가 출력

y=14
print(y) # 14

x = 100
print(x) # 100

```

*변수의 이름을 정하는 규칙*

- 반드시 문자 또는 underscore(_)로 시작합니다.(숫자로 시작 불가)

- 문자와 숫자 underscore(_)포함 가능합니다.

- 읽는 사람이 읽기 편하게 변수명을 정합니다.

```

#Case1-Worst

xiojqwerlkj = 35.0
dpemdnlem = 12.50
rlaxotjs = xiojqwerlkj * dpemdnlem

print(rlaxotjs) #437.5


#Case2 - Bad

a = 35.0

b = 12.50

c = a* b

print(c) #437.5 

#Case30 Good

hours = 35.0
rate = 12.50

pay = hours * rate

print(pay)#437.5
```

**할당문(Assignment Statements)**

대입문은 오른쪽 표현의 결과를 왼쪽 변수에 저장하는 걸로 구성합니다.

```
x = 0.6

x = 3.9*x*(1-x)
print(x) # 0.936
      #3.9*0.6*(1-0.6)
      
      
x = 3.9* x * (1-x)
print(x) # 0.2336256

      # 3.9*0.936*.(1-0.936)


```
