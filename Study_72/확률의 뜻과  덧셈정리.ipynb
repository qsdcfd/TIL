## 확률

### 이론적 확률

- 정말 눈 혹은 머리로 생각할 수 있는 직관적인 확률



### 경우의 수를 이용한 확률

- 실제 벌어진 여러 사건들을 나열하여 특정 경우의 수를구하는 것 보통 수식이나 그림으로 표현이 자주 된다



### 통계적 확률

- 어떠한 사건이 보고 싶은 확률/ 전체 확률 을 나눠서 생각하는 확률로 보통은 사건을 수치화를 통해서 확률을 구한다. 즉 어떤 사건을 n번 반복하여 일어나는 횟수가 있고 일어나는 횟수 / 전체를 사건이 일어날 통계적 확률이다.




### 이론적 확률과 통계적 확률

- 이론적 확률: 눈을 감고 직관적으로 생각

- 통계적 확률: 여러번 뽑아서 어떠한 확률로 수렴하게 하는 거



### 확률의 덧셈법칙

- 전체사건: N

- 특정사건: A

- 특정 사건에 대한 확률 A / N




### 벤다이어 그램과 확률

- 벤다이어그램으로 확률을 표현한 것이다
