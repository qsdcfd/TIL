## 단층 퍼셉트론으로 게이트 연산 구현하기

- 단층 퍼셉트론 식 이용

$$ y = x * w+b$$

- 입력값: x

- 가중치: w

- 편향: b

### 식 설명

- 각각의 입력값에는 각각의 가중치가 존재하는데, 가중치의 값이 클수록 해당 입력값이 중요하다는 것을 의미합니다.

- 각 입력값이 가중치와 곱해져서 인공 뉴런에 보내지고, 각 입력값과 그에 해당되는 가중치의 곱의 전체 합이 임계치를 넘으면 인공 뉴런은 출력 신호로 1을 출력하고 그렇지 않으면0 을 출력

image.png

- 편향인 b는 딥러닝의 최적의 값을 찾기 위한 변수이고 뉴런이 얼마나 쉽게 활성화가 되는지를 결정하는 것입니다.

   - 예를 들어, b = -1이라면 w1*x1은 값은 1을 넘어야만 활성화가 됩니다. 근데 b =-100이라면 100이상의 값을 넘겨야합니다.

image.png

## 게이트 연산 구현

```python
def AND(x1, x2):
    w1,w2, b = 0.5,0.5, 0.7
    act = x1*w1 + x2*w2

    if act <= b:
        return 0
    elif act > b:
        return 1


def OR(x1, x2):
    w1,w2, b = 0.5,0.5, 0.4
    act = x1*w1 + x2*w2

    if act <= b:
        return 0
    elif act > b:
        return 1

def NAND(x1, x2):
    w1,w2, b = -0.5,-0.5, -0.7
    act = x1*w1 + x2*w2

    if act <= b:
        return 0
    elif act > b:
        return 1

def XOR(x1, x2):
    A = OR(x1,x2)
    B = NAND(x1, x2)
    return AND(A,B)
```
