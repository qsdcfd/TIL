# 벡터공간(Vector Spaces)


## 개론

- 벡터: 크기와 방향을 모두 가진 물리량으로 보통 화살표로 표현합니다.

- 벡터의 크기: 화살표의 길이

- 벡터의 방향: 화살표의 방향

- 크기와 방향이 같다면 같은 벡터입니다.



- 합: 두 물리량이 결합될 때 나타나는 효과는 두 벡터를 결합시켜 얻은 합성벡터입니다.

- 평생사변형 법칙: 두 벡터를 결합시키는 규칙



*시점이 O로 일치하는 두 벡터 a,b의 합은 점 O에서 시작하는 벡터이고, a와b를 이웃한 변으로 하는 평행사변형의 대각선으로 나타나게 됩니다.*



*벡터의 합과 벡터의 스칼라 곱*

- 벡터의 스칼라 곱은 벡터의 크기에 실수를 곱하면 확대하거나 축소가 가능합니다.

   
- 벡터의 평행: 두 벡터의 방향이 같거나 180도 반대인 벡터들을 말합니다.


- 벡터의 대수적 성질

*훗날, 직선의 방정식과 평면의 방정식을 나타내는 데 사용이 됩니다.

*벡터 공간 혹은 선형공간에서 이뤄지는 것입니다.

> - 모든 벡터 x,y에 대하여 x + y = y + x이다(교환법칙 성립)                                                                                                                         
> - 모든 벡터 x,y,z에 대하여 (x+y) +z = x + (y+x)이다. (결합법칙 성립)
> - 모든 벡터 x에 대하여 x+ 0  = x를 만족하는 벡터 0이 존재한다.
> - 각 벡터 x마다 x+y=0을 만족하는 벡터 y가 존재한다.
> - 모든 벡터 x에 대하여 1x=x이다.
> - 모든 실수 a,b와 모든 벡터 x에 대하여 (ab)x = a(bx)이다.
> - 모든 실수 a와 모든 벡터 x,y에 대하여 a(x+y) = ax +ay이다.
> - 모든 실수 a,b와 모든 벡터 x에 대하여 (a+b)x = ax+bx이다.

## 벡터공간

여기서는 벡터공간의 원소를 말할 것이고 V(벡터공간)을 뜻합니다.

- n순서쌍(n-tuple): 공간 F의 원소 (a1,a2,a3,...an)에서 성분을 가져온다는 의미입니다.

- n순서쌍의 성분: a1,a2,a3,..an입니다.

- 같은 공간에서 성분을 가져온 두 순서쌍(a1,a2,....an)과 (b1,b2,...bn)이 ai=bi(i=1,2,3,..n)일 때 같다고 정의합니다.



- F^n의 벡터는 열벡터로 표현합니다.




- m x n행렬




*aij (1<=i<=m, 1<=j<=n)는 F의 원소입니다.

*대각성분 : 성분 aij ( i=j)

*행(row): 행렬의 i번째 , F^n의 벡터, 1 x n

*열(column): 행렬의 j번째, F^m의 벡터, m x 1



*영행렬: 모든 성분이 0인 m x n입니다



*정사각행렬: 행의 개수와 열의 개수가 같은 행렬입니다.



*같다: 두 mxn행렬 A,B에 대응하는 성분이 모두 일치할 경우(Aij = Bij)


- 다항식: 여러 항들이 모여있는 것을 말합니다.



*degree: 계수가 0이 아닌 항의 지수 중 가장 큰 값으로 정의합니다. 만약, degree=0이면 f(x)=c꼴(c는상수)이다.

- 벡터 합의 소거 법칙

*조건: x,y,z는 벡터공간이고 x + z = y + z일때, x=y이다.

- 모든 벡터공간V에 대해 성립합니다.

*모든 벡터 x에 대하여 0x=0이다.

*모든 스칼라 a와 모든 벡터 x에 대하여 (-a)x=-(ax)=a(-x)이다.

*모든 스칼라 a에 대하여 a0 =0이다.
