```python
import sys
sys.setrecursionlimit(10**6)
input = sys.stdin.readline

N = int(input())
M = int(input())
parent = [0] * (N+1) 
rank = [0] * (N+1) 
edges = [[] for i in range(M+1)] 


for i in range(1, N+1):
    parent[i]=i


for i in range(1, M+1): 
    a, b, c = map(int,input().split())
    edges[i].extend([c, a, b])


def find(a): 
    if parent[a] == a: 
        return a
    p = find(parent[a]) 
    parent[a] = p 
    return parent[a]

def union(a, b): 
    a = find(a) 
    b = find(b)
    if a == b:
        return
    if rank[a] > rank[b]:  
        parent[b] = a
    else:
        parent[a] = b
        if rank[a] == rank[b]: 
            rank[b] += 1 


def kruskal(edges):
    edges.sort() 
    total = 0 
    MST = [] 
    
    for edge in edges:
        if not edge: 
            continue
        cost, a, b = edge 
        if find(a) != find(b): 
            union(a,b)
            total += cost 
            MST.append((a,b)) 

    return total, MST

total, MST = kruskal(edges)
print(total)
```
