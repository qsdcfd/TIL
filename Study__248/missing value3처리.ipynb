```python
import os
import gc
import warnings

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data= pd.read_csv('../Downloads/파이널프로젝트_RAW_210329_210926.csv')
data.head(10)
#105419 rows X 18 columns
```

```python
data.info()
```

```python
data['coupon_discount_amount'] = data['sale_price'] - data['transaction_amount']#coupon_discount결측치 제거
```

```python
data['sale_price'] = data['sale_price'].fillna(data['transaction_amount']) #sale_price 결측치 제거
```

```python
del data['tax_free_amount']
```

```python
data['pg'] = data['pg'].fillna('default') #NaN은 초기값이므로 default
```

```python
data['marketing_start_at'] = data['marketing_start_at'].fillna(data['completed_at']) #날짜 없는 것은 마지막 끝 날짜로
```

```python
#method 결측치를 위한 최빈값 구하기
data['method'].agg(**{
    'most_common_value':lambda x:x.mode()
}).reset_index()
```

```python

data['method'] = data['method'].fillna('CARD') #최반값 설정
```

```python
idx = data[data['type'] == "ORDER" ].index     
data.drop(idx , inplace=True) # 해당 인덱스를 제거함
```

```python
idx = data[data['type'] == "TRANSACTION" ].index     
data.drop(idx , inplace=True) # 해당 인덱스를 제거함
```

```python
idx = data[data['state'] == "PENDING"].index     
data.drop(idx , inplace=True) # 해당 인덱스를 제거함
```

```python
idx = data[data['state'] == "CANCELLED"].index     
data.drop(idx , inplace=True) # 해당 인덱스를 제거함
```

```python
idx = data[data['state'] == "HIDDEN"].index     
data.drop(idx , inplace=True) # 해당 인덱스를 제거함
```

```python
idx = data[data['state'] == "DELETED"].index     
data.drop(idx , inplace=True) # 해당 인덱스를 제거함
```
